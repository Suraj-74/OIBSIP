## Unemployment in India during Covid-19 dataset analysis by Suraj Yadav

### The story behind this datasets is how lock-down affects employment opportunities and how the unemployment rate increases during the Covid-19.

This dataset contains the unemployment rate of all the states in India

Region = states in India

Date = date which the unemployment rate observed

Frequency = measuring frequency (Monthly)

Estimated Unemployment Rate (%) = percentage of people unemployed in each States of India

Estimated Employed = percentage of people employed

Estimated Labour Participation Rate (%) = labour force participation rate by dividing the number of people actively<br>
participating in the labour force by the total number of people eligible to participate in the labor force

## Graphical Analysis

```python
# plot a boxplot to check outliers
df.plot(kind="box", subplots=True, layout=(5,2), figsize=(14,11))
plt.show()
```

```python
df.hist(figsize=(12,11))
```

```python
# Statewise umenployment rate
plt.figure(figsize=(12,12))
sns.countplot(x=df["Region"], palette="plasma")
plt.xticks(rotation=90)
plt.show()
```

```python
# plot a pie chart for area column
a=df["Area"].value_counts()
plt.figure(figsize=(10,9))
plt.pie(x=a.values, labels=a.index, autopct="%.2f%%")
plt.show()
```

```python
# analysis by year
a1=df["Year"].value_counts()
a1
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(11,10))
plt.bar(a1.index, a1.values,color="Cyan")
plt.title("Unemployment rate per year")
plt.xticks(rotation=90)
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
## unemployment rate per month
sns.set_style("darkgrid")
plt.figure(figsize=(10,9))
sns.countplot(df["Month"])
plt.title("Unemployment rat per month")
plt.xlabel("Month")
plt.ylabel("Count")
plt.show()
```

```python
# unemployment rate per day
sns.set_style("whitegrid")
plt.figure(figsize=(10,9))
sns.countplot(df["Day"], palette="dark")
plt.title("Unemployment ray per day")
plt.xlabel("Day")
plt.ylabel("count")
plt.show()
```

```python
b1=df.groupby("Region")["Estimated Unemployment Rate(%)"].mean()
b2=b1.sort_values(ascending=True)
print(b2)
plt.figure(figsize=(15,7))
sns.lineplot(x=b1.index,y=b1.values,marker="*",color="maroon")
plt.title("Region vs Estimated Unemployment Rate(%)")
plt.xlabel("Regions")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
a2=df.groupby("Region")["Estimated Employed"].mean()
a3=a2.sort_values(ascending=True)
print(a3)
plt.figure(figsize=(11,10))
sns.lineplot(x=a2.index,y=a2.values,marker="o",color="blue")
plt.title("Region vs Estimated Employed'")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
sns.set_style("darkgrid")
a=df.groupby("Region")["Estimated Labour Participation Rate(%)"].mean()
a1=a.sort_values(ascending=True)
print(a1)
plt.figure(figsize=(10,9))
sns.lineplot(x=a.index,y=a.values, marker="o", color="pink")
plt.title("Region vs Estimated Labour Participation Rate(%)")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```
