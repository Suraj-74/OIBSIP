## Select independent and dependent variables

```python
x=df.drop("Sales", axis=1)
y=df["Sales"]
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
```

## Splitting the data into training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
print(198*0.2)
print(198-39.6)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Import the model

## Linear Regression

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
def reg_eval(y_test,ypred):
    mae=mean_absolute_error(y_test,ypred)
    mse=mean_squared_error(y_test,ypred)
    rmse=np.sqrt(mean_squared_error(y_test,ypred))
    r2=r2_score(y_test,ypred)
    print("mae:",mae)
    print("mse:",mse)
    print("rmse:",rmse)
    print("r2_score",r2)
    
def train_test_score(model):
    print("Training Score:", model.score(x_train,y_train))
    print("Testing Score:", model.score(x_test,y_test))
```

```python
leg_reg=LinearRegression()
leg_reg.fit(x_train,y_train)
```

```python
ypred_legreg=leg_reg.predict(x_test)
print(ypred_legreg)
```

## Evaluate the model

```python
reg_eval(y_test,ypred_legreg)
```

```python
train_test_score(leg_reg)
```

```python
r2_lr=0.8989454779619588
```

from linear regression we got 89% traing and testing score and r2 score is also 89%

## KNN-Regression

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
```

```python
ypred_knn=knn.predict(x_test)
print(ypred_knn)
```

## Evaluate the model

```python
reg_eval(y_test,ypred_knn)
```

```python
train_test_score(knn)
```

```python
r2_knn=0.9260560186704322
```

Here, we can see that r2 score of knn model is greater than log_reg model,
So our best model is KNN-Regression

```python
final=pd.DataFrame({"Models":["LR","KNN"],"R2_Score":[r2_lr,r2_knn]})
```
